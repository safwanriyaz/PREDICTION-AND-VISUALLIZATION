```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import yfinance as yf

dataset = 'AAPL'
data = yf.download(dataset, start='2016-02-01', end='2018-04-02')

time_series = data['Close']

train_size = int(len(time_series) * 0.8)
train, test = time_series[:train_size], time_series[train_size:]

model = ARIMA(train, order=(5, 1, 2))
model = model.fit()

y_pred = model.forecast(steps=len(test))

mse = mean_squared_error(test, y_pred)
print(f'Mean Squared Error: {mse:.4f}')

plt.figure(figsize=(10, 6))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test', color='orange')
plt.plot(test.index, y_pred, label='Predicted', color='green')
plt.title(f'Time Series Forecasting for {dataset}')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
#catboost
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris=datasets.load_iris()
x=iris.data
y=iris.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.6,random_state=50)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)

cbm=CatBoostClassifier(iterations=1000,learning_rate=0.6,depth=6,verbose=False)
model=cbm.fit(x_train,y_train)
y_pred=cbm.predict(x_test)

accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy:{'accuracy.3f'}')
print(accuracy_score)

feature_importance=cbm.get_feature_importance()
plt.bar(range(len(feature_importance)),feature_importance,color='red')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend('catboost_info')
plt.grid(color='green')
plt.show()
```

```python
#catboost
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,classification_report

df=pd.read_csv("C:\ds\Datasets\Indian_earthquake_data.csv")
print(df.to_string)
x=df[["Magnitude"]]
y=df["Depth"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.6,random_state=50)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)

cbm=CatBoostClassifier(iterations=10,max_depth=6,random_state=50)
model=cbm.fit(x_train,y_train)
y_pred=cbm.predict(x_test)

mse=mean_squared_error(y_test,y_pred)
print(mse)

plt.scatter(x_train,y_train,color='red')
plt.xlabel('TRAINED DATA')
plt.legend('catboost_info')
plt.grid(color='green')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix

df=pd.read_csv("C:\ds\Datasets\Retail_sales.csv")
print(df.head())

x=df[["Product ID"]]
y=df["Units Sold"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)

model=LogisticRegression(random_state=42)

model.fit(x_train,y_train)

y_pred=model.predict(x_test)

accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy: {accuracy:2f}')

plt.figure(figsize=(10,6))
plt.bar(y_test,y_pred)
plt.show()
            
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

df=pd.read_csv("C:\ds\Datasets\Retail_sales.csv")
print(df.to_string)
print(df.head())

x=df[["Product ID"]]
y=df["Units Sold"]


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=42)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)

knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)

accuracy=accuracy_score(y_test,y_pred)
conf_matrix=confusion_matrix(y_test,y_pred)
class_report=classification_report(y_test,y_pred)

print(f'Accuracy: {accuracy:.2f}')
print('Confusion Matrix:')
print(conf_matrix)
print('classification_report:')
print(class_report)


plt.scatter(x,y,c=y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sales',color='green',fontsize=10, fontweight='bold', style='italic')
plt.ylabel('Progress',color='red',fontsize=10, fontweight='bold', style='italic')
plt.title('Retail sales classification',fontsize=14, fontweight='bold')
plt.legend()
plt.grid(color='black',linestyle='dotted')
plt.show()

```
